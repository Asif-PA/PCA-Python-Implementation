```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_df=load_breast_cancer()
```

```python
cancer_df.keys()
```

```python
print(cancer_df['DESCR'])
```

```python
df=pd.DataFrame(cancer_df['data'],columns=cancer_df['feature_names'])
df.head()
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['concavity error'],df['fractal dimension error'],color='red',marker='*',s=50)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar=StandardScaler()
```

```python
scalar.fit(df)
```

```python
scaled_data=scalar.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
print(x_pca)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0],x_pca[:,1],marker="*",s=30,c=cancer_df['target'],cmap="plasma")
plt.title("Scatter Plot After PCA")
plt.xlabel("1st Principal component")
plt.ylabel("2nd Principal component")
```
